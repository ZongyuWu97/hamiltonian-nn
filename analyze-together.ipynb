```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and axes
fig, ax = plt.subplots()

# Function to draw the spring
def draw_spring(x_start, y_start, x_end, y_end, coils=10, width=0.1):
    length = y_start - y_end
    coil_height = length / coils
    coil_width = width / 2
    x = []
    y = []
    for i in range(coils + 1):
        x.append(x_start + coil_width * (-1)**i)
        y.append(y_start - i * coil_height)
    ax.plot(x, y, color='black', linewidth=3)

# Draw the ceiling
ax.plot([0.3, 0.7], [0.8, 0.8], color='black', linewidth=3)

# Draw the spring
draw_spring(0.5, 0.8, 0.5, 0.4)

# Draw the mass
mass = plt.Rectangle((0.45, 0.2), 0.1, 0.2, edgecolor='black', facecolor='none', linewidth=3)
ax.add_patch(mass)

# Add labels
ax.text(0.5, 0.25, '$m$', fontsize=20, ha='center')
ax.text(0.55, 0.6, '$q$', fontsize=20, va='center')
ax.text(0.65, 0.3, '$p = m\dot{q}$', fontsize=20, va='center')

# Add the arrow
ax.annotate('', xy=(0.6, 0.8), xytext=(0.6, 0.4),
            arrowprops=dict(arrowstyle='<->', linewidth=1.5))

# Set limits and hide axes
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

# Show plot
plt.tight_layout()

plt.show()
fig.savefig('mass_spring.png')
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Create figure and axes
fig, ax = plt.subplots()

# Draw the pivot (ceiling)
ax.plot([0.3, 0.7], [0.9, 0.9], color='black', linewidth=3)

# Draw the vertical dotted line
ax.plot([0.5, 0.5], [0.9, 0.2], color='black', linewidth=1, linestyle='dotted')

# Draw the string of the pendulum
string_length = 0.5
angle = 30  # angle in degrees

# Calculate the end of the string coordinates
x_end = 0.5 + string_length * np.sin(np.radians(angle))
y_end = 0.9 - string_length * np.cos(np.radians(angle))

# Draw the string
ax.plot([0.5, x_end], [0.9, y_end], color='black', linewidth=3)

# Draw the bob of the pendulum
bob = patches.Circle((x_end, y_end), 0.05, edgecolor='black', facecolor='none', linewidth=3)
ax.add_patch(bob)

# Draw the angle arc
arc = patches.Arc((0.5, 0.9), 0.3, 0.3, angle=0, theta1=270, theta2=270+angle, linewidth=1.5, linestyle='dashed')
ax.add_patch(arc)

# Add labels
ax.text(0.55, 0.65, '$q$', fontsize=20, ha='center')
ax.text(0.7, 0.3, '$p = m\\dot{q}$', fontsize=20, va='center')

# Draw an arrow for the angle Î¸
ax.annotate('', xy=(0.5 + 0.15 * np.cos(np.radians(270+angle/2)), 0.9 - 0.15 * np.sin(np.radians(270+angle/2))),
            xytext=(0.5, 0.9),
            arrowprops=dict(arrowstyle='<-', linewidth=1.5))

# Set limits and hide axes
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

# Show plot
plt.show()
fig.savefig('ideal_pendulum.png')
```
